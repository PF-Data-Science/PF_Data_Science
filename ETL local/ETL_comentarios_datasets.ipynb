-**user_id**: Es el ID del usuario que dejó la reseña.  
-**name**: Es el  nombre del usuario que dejó la reseña.  
-**time**: La fecha y hora en que se dejó la reseña.  
-**rating**: La calificación que el usuario le dio al lugar, en una escala de 1 a 5 estrellas.  
-**text**: El texto de la reseña.  
-**pics**: Una lista de imágenes adjuntas a la reseña.  
-**resp**: La respuesta del propietario del negocio a la reseña.  
-**gmap_id**: Es el ID de Google Maps del lugar que se revisó.

```python
# TRATAMIENTO DE VALORES DEL DATAFRAME...

# Identificar valores nulos en el DataFrame y contarlos por columna
valores_nulos = reseñas_googlemaps.isnull().sum()

# Identificar valores faltantes (en blanco) en el DataFrame y contarlos por columna
valores_faltantes = (reseñas_googlemaps == '').sum()

# Identificar valores iguales a "SD" en el DataFrame y contarlos por columna
valores_SD = (reseñas_googlemaps == 'SD').sum()

# Identificar valores iguales a "sd" en el DataFrame y contarlos por columna
valores_sd = (reseñas_googlemaps == 'sd').sum()

# Calcular porcentajes
porcentaje_nulos = (valores_nulos / len(reseñas_googlemaps)) * 100
porcentaje_faltantes = (valores_faltantes / len(reseñas_googlemaps)) * 100
porcentaje_SD = (valores_SD / len(reseñas_googlemaps)) * 100
porcentaje_sd = (valores_sd / len(reseñas_googlemaps)) * 100

# Crear un DataFrame con los resultados
resultados = pd.DataFrame({
    'Valores Nulos': valores_nulos,
    'Porcentaje Nulos': porcentaje_nulos,
    'Valores Faltantes': valores_faltantes,
    'Porcentaje Faltantes': porcentaje_faltantes,
    'Valores "SD"': valores_SD,
    'Porcentaje "SD"': porcentaje_SD,
    'Valores "sd"': valores_sd,
    'Porcentaje "sd"': porcentaje_sd
})
```
